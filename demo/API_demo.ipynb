```python
from skimage.io import imread, imshow
from pyclesperanto import Clesperanto

cle = Clesperanto()
cle
```

```python
image = imread("blobs.tif")

imshow(image)
```

```python
blurred = cle.create(image.shape)
cle.gaussian_blur(cle.push(image), blurred, 3.0, 3.0, 0.0)

imshow(cle.pull(blurred))
```
