## Image API

In this notebook we want to have a quick look at the pyclesperanto Image API and its possible interoperability with other python libraries.

```python
from skimage.io import imread, imshow
import numpy as np
import pyclesperanto as cle
```

### Numpy images

```python
image = imread("https://samples.fiji.sc/blobs.png").squeeze()
```

```python
print(type(image), image.ndim, image.shape, image.dtype)
```

```python
image
```

```python
print(image)
```

```python
imshow(image)
```

### Pyclesperanto images

```python
img = cle.asarray(image)
```

```python
print(type(img), img.ndim, img.shape, img.dtype)
```

```python
img
```

```python
print(img)
```

```python
cle.imshow(img)
```
