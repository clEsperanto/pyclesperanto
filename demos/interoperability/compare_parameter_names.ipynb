# Check names of parameters
In this notebook we are comparing parameter names between pyclesperanto and its _prototype. Goal is to make backwards compatibility breaking changes minimal.

```python
import pyclesperanto as cle
cle.__version__
```

```python
import pyclesperanto_prototype as clep
clep.__version__
```

```python
import bia_bob
import inspect
```

```python
cle_functions = cle.operations()
len(cle_functions), type(cle_functions)
```

```python
clep_functions = clep.operations()
len(clep_functions), type(clep_functions)
```

```python
# Retrieve keys that are common to both dictionaries
common_keys = list(set(cle_functions.keys()).intersection(clep_functions.keys()))
len(common_keys), common_keys[:3]
```

```python
def in_name_changes_that_are_ok(cle_name, clep_name):
    return ((cle_name == "input_image" and clep_name == "source") or 
            (cle_name == "output_image" and clep_name == "destination") or 
            (cle_name == "input_image" and clep_name == "labels_input") or
            (cle_name == "output_image" and clep_name == "labels_destination") or 
            (cle_name == "input_image0" and clep_name == "source1") or 
            (cle_name == "input_image0" and clep_name == "image1") or 
            (cle_name == "input_image1" and clep_name == "image2") or           
        False)
```

```python
def compare_function(function_name):

    cle_sample_function = getattr(cle, function_name)
    cle_parameter_names = [p.name for p in inspect.signature(cle_sample_function).parameters.values()]

    clep_sample_function = getattr(clep, function_name)
    clep_parameter_names = [p.name for p in inspect.signature(clep_sample_function).parameters.values()]

    if "device" in cle_parameter_names:
        cle_parameter_names.remove("device")
    else:
        print(f"Function {function_name} in pyclesperanto has no device parameter")

    if len(cle_parameter_names) != len(clep_parameter_names):
        print(f"Function {function_name} has diverging number of parameters:", 
              len(cle_parameter_names), 
              len(clep_parameter_names), 
              cle_parameter_names, 
              clep_parameter_names)
    
    for cle_name, clep_name in zip(cle_parameter_names, clep_parameter_names):
        if cle_name != clep_name and not in_name_changes_that_are_ok(cle_name, clep_name):
            print(f"Function {function_name} has diverging parameter names: {cle_name} != {clep_name}")


compare_function(common_keys[0])
```

```python
for function_name in common_keys:
    compare_function(function_name)
```
