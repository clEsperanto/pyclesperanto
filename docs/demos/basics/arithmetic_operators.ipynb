# Array Arithmetic Operations

pyclesperanto provides several functions for arithmetic operations between arrays and constants, or between arrays.

These operations are element-wise and run very efficiently on GPUs as they are easily parallelized.

````python
import pyclesperanto as cle
import numpy as np

cle.get_device() # default device initialisation
````

## Operations with Scalars

Operations between arrays and scalars (constants) can be performed using operators or equivalent functions:

**Arithmetic operations:**
- `array * x` → `cle.multiply_image_and_scalar()`
- `array / x` → `cle.divide_image_and_scalar()`
- `array + x` → `cle.add_image_and_scalar()`
- `array - x` → `cle.subtract_image_from_scalar()`

**Comparison operators:**
- `array > x` → `cle.greater_constant()`
- `array >= x` → `cle.greater_or_equal_constant()`
- `array < x` → `cle.smaller_constant()`
- `array <= x` → `cle.smaller_or_equal_constant()`
- `array == x` → `cle.equal_constant()`

These operations make it easy to compute mathematical equations involving arrays and constants.

````python
# Create sample array
array = cle.push(np.arange(1, 11, dtype=float))

# Arithmetic operations with scalar
result_mult = array * 2
result_div = array / 5
result_add = array + 10
result_sub = array - 100

print("Original array:", array)
print("Array * 2:", result_mult)
print("Array / 5:", result_div)
print("Array + 10:", result_add)
print("Array - 100:", result_sub)

# Comparison with scalar
mask_greater = array > 5
mask_equal = array == 5

print("\nArray > 5:", mask_greater)
print("Array == 5:", mask_equal)
````

## Operations Between Arrays

Similar operations are available between arrays, with the requirement that they must have the same shape.

**Arithmetic operations:**
- `array * array` → `cle.multiply_images()`
- `array / array` → `cle.divide_images()`
- `array + array` → `cle.add_images()`
- `array - array` → `cle.subtract_images()`

**Comparison operators:**
- `array > array` → `cle.greater()`
- `array >= array` → `cle.greater_or_equal()`
- `array < array` → `cle.smaller()`
- `array <= array` → `cle.smaller_or_equal()`
- `array == array` → `cle.equal()`

````python
# Create two sample arrays
array_a = cle.push(np.array([1, 2, 3, 4, 5], dtype=float))
array_b = cle.push(np.array([2, 3, 1, 5, 2], dtype=float))

# Arithmetic operations between arrays
result_mult = array_a * array_b
result_div = array_a / array_b
result_add = array_a + array_b
result_sub = array_a - array_b

print("Array A:", array_a)
print("Array B:", array_b)
print("A * B:", result_mult)
print("A / B:", result_div)
print("A + B:", result_add)
print("A - B:", result_sub)

# Comparison between arrays
mask_greater = array_a > array_b
mask_equal = array_a == array_b

print("\nA > B:", mask_greater)
print("A == B:", mask_equal)
````

### Performance Optimization with `evaluate()`

Array operators (e.g., `+`, `*`) and their associated functions (e.g., `multiply_images()`, `add_image_and_scalar()`) generate and execute one GPU kernel per operation.

When performing complex arithmetic equations like:

```
f = (cos(A)² + sin(B)²) / C
```

Multiple kernels are executed in sequence, which can be inefficient compared to executing a single kernel for the entire computation.

For such cases, pyclesperanto provides the `evaluate()` function, which allows you to pass an equation directly and has it compiled into a single kernel:

**Note:** Some operators are not available in `evaluate()`, such as `^` or `**`. Use the function equivalents instead, e.g., `pow()`, `cos()`, `sin()`, `log()`.

````python
import time

# Create test arrays
array_A = cle.push(np.ones((1000, 1000)))
array_B = cle.push(np.ones((1000, 1000)))
constant = 10

# Method 1: Using multiple operations (multiple kernels)
start = time.time()
result_multi = (cle.power(cle.cos(array_A), scalar=2.0) + cle.power(cle.sin(array_B), scalar=2.0)) * constant
end = time.time()
multi_kernel_time = end - start
print(f"Multiple kernels: {multi_kernel_time:.6f} seconds")

# Method 2: Using evaluate (single kernel)
start = time.time()
result_single = cle.evaluate("(pow(cos(a), 2) + pow(sin(b), 2)) * c", 
                              {"a": array_A, "b": array_B, "c": constant})
end = time.time()
one_kernel_time = end - start
print(f"Single kernel (evaluate): {one_kernel_time:.6f} seconds")

# Compare performance
speedup = multi_kernel_time / one_kernel_time
print(f"\nevaluate() is {speedup:.1f}x faster")
````
